   Evaluate the derivative at x = 1.
   
   

**Your answer here in LaTex syntax:**

- plot the function and its derivative using matplotlib.

```python
# your code here
def f(x):
    return np.tanh(3*x**3 - x)
def f_prime(x):
    return (1 - np.tanh(3*x**3 - x)**2) * (9*x**2 - 1)
x_vals = np.linspace(-1, 1, 1000)
y_vals = f(x_vals)
y_prime_vals = f_prime(x_vals)

plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(x_vals, y_prime_vals, label="f'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
